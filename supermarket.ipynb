```python
import pandas as pd 
import seaborn as se 
data = pd.read_csv('/home/kali/Downloads/supermarket_sales - Sheet1.csv')
data.head(1)
```

```python
df =data.pivot_table(index = ['Gender','Customer type'],columns = 'City',values='Total',aggfunc = 'sum',margins = False,sort=True,observed= True)
df
```

```python
import matplotlib.pyplot as plt
col = df.columns
a = df[[col[i] for i in range(df.shape[1])]].plot(kind ='bar',subplots=False,legend = 'reverse').legend(bbox_to_anchor=(1,1))
```
